```python
import numpy as np
import pandas as pd

#Use pandas to read data into pd frame
data = pd.read_csv('https://raw.githubusercontent.com/Micniks/Python-Week11-Group-3-Assignments/main/cultists.csv')
print(data.head())
```

####  Divide the data into two datasets, cultists and roles. Make the X dataset for  the stats(features) of the cultists, and the Y dataset for the position(targets) for the dataset.
#### Refactor the dataset cultists, by removing the irrelevent features from in the dataset, so it only contains charactistic statistics
 ['stealth', 'Influence', 'Endurance', 'Lore', 'Economic', 'Strength', 'Insanity']

#### Refactor the dataset roles so each value become machine friendly numbers, instead of strings.

```python
#Dropping unrelated columns (features/observations)

X = np.array(data.drop(columns=['Cultist','Position','Age','Membership_in_years','Contribution_in_dollars','Members_recruited','Type_of_living_area']))

#getting the response data
y = data[['Position']]

#converting the y data (labels) into machine friendly numbers
label_mapping = {
    'Priest':0, 
    'Enforcer':1, 
    'Assassin':2, 
    'Recruiter':3,
    'Accountant':4,
    'Advisor':5
}

print('Before converting to machine friendly numbers: \n',y[:5])
y['Position'] = y['Position'].map(label_mapping)

y = np.array(y)

print('After converting:\n',y[:5])

print(X.shape)
print(y.shape)
```

#### Use LinearRegression to make a model object.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

model = LinearRegression()
```

#### Use train_test_split from model_selection, to make training and testing dataset from cultists and roles

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

model.fit(X_train,y_train)
```

#### Feed the model with the training datasets, and score the model on the training dataset (model.score)
#### Score the model on the testing datasets, and compare to the testing scores.

```python
y_pred = model.predict(X_test)
print(len(y_pred))
```

```python
result_test = model.score(X_test, y_test)
print(result_test)

result_train = model.score(X_train, y_train)
print(result_train)
```

```python
X1 = data.drop(columns=['Cultist','Position'])
X1.head()
```

#### Change the Living_Area feature into multiple numeric features, using One Hot Encoding

```python
from sklearn import preprocessing

label_enc =preprocessing.LabelEncoder()
X1 = pd.get_dummies(X1,columns=['Type_of_living_area'])
X1.head()


```

#### Remove all rows with missing data for any features

```python
print('rows before drop n/a',len(X1))
X1_filtered = X1.dropna()
print('rows after',len(X1_filtered))
```

#### Make a model/analyzer from sklearn.cluster, with the appropriate bandwidth for the data.
#### Use the analyzer to process the data, grouping people into clusters.

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
analyzer = MeanShift(bandwidth=550) 

analyzer.fit(X1_filtered)

labels = analyzer.labels_
X1_filtered['cluster_group'] = np.nan
data_length=len(X1_filtered)
for i in range(data_length): 
    X1_filtered.iloc[i,X1_filtered.columns.get_loc('cluster_group')] = labels[i] #set the cluster label on each row

X1_filtered
```

#### Make a dataset from the cluster-array, showing the avarage statistic values for features in each cluster
#### Add a count feature to each cluster, showing how many members are in each cluster

```python
#Grouping passengers by Cluster
X1_cluster_data = X1_filtered.groupby(['cluster_group']).mean()
#Count of passengers in each cluster
X1_cluster_data['Counts'] = pd.Series(X1_filtered.groupby(['cluster_group']).size())
print(X1_cluster_data)
```
